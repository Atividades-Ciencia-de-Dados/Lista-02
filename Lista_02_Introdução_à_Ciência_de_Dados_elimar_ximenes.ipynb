##e) Utlizando o DuckDB recupere:

```python
def executar_consulta(query):
    con = duckdb.connect('df_banco.duckdb')
    resultado = con.execute(query).fetchdf()
    con.close()
    return resultado
```

###19. Mensagens idênticas compartilhadas pelo mesmo usuário (e suas quantidades);

```python
# Retorna mensagens idênticas compartilhadas pelo mesmo usuário e suas quantidades
mensagens_repetidas_usuario = executar_consulta("""
    SELECT
        id_member_anonymous,
        text_content_anonymous,
        COUNT(*) AS quantidade_repeticoes
    FROM fake_telegram
    WHERE
        text_content_anonymous IS NOT NULL
        AND id_member_anonymous IS NOT NULL
    GROUP BY
        id_member_anonymous, text_content_anonymous
    HAVING
        COUNT(*) > 1
    ORDER BY
        quantidade_repeticoes DESC
""")

mensagens_repetidas_usuario
```

###20. Mensagens idênticas compartilhadas pelo mesmo usuário em grupos distintos (e suas quantidades);

```python
# Retorna mensagens idênticas compartilhadas pelo mesmo usuário em grupos distintos e suas quantidades
mensagens_repetidas_usuario_grupos = executar_consulta("""
    SELECT
        id_member_anonymous,
        text_content_anonymous,
        COUNT(DISTINCT id_group_anonymous) AS quantidade_grupos
    FROM fake_telegram
    WHERE
        text_content_anonymous IS NOT NULL
        AND id_member_anonymous IS NOT NULL
        AND id_group_anonymous IS NOT NULL
    GROUP BY
        id_member_anonymous, text_content_anonymous
    HAVING
        COUNT(DISTINCT id_group_anonymous) > 1
    ORDER BY
        quantidade_grupos DESC
""")

mensagens_repetidas_usuario_grupos
```

###21. Os 30 unigramas, bigramas e trigramas mais compartilhados;

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import nltk
from nltk.corpus import stopwords

# Baixando stopwords do NLTK
nltk.download('stopwords')

# Função auxiliar para extrair os top n n-gramas
def top_ngrams(texts, ngram_range=(1,1), top_n=30):
    stop_words = stopwords.words('portuguese')  # Lista de stopwords em português
    vectorizer = CountVectorizer(ngram_range=ngram_range, lowercase=True, stop_words=stop_words)
    X = vectorizer.fit_transform(texts)
    total_counts = X.sum(axis=0).A1
    vocab = vectorizer.get_feature_names_out()
    counts = pd.Series(total_counts, index=vocab)
    return counts.sort_values(ascending=False).head(top_n)

# Usando sua função `executar_consulta` para obter os textos
textos = executar_consulta("""
    SELECT text_content_anonymous FROM fake_telegram WHERE text_content_anonymous IS NOT NULL
""")

# Unigramas
top_unigrams = top_ngrams(textos['text_content_anonymous'], ngram_range=(1,1))
print("Top 30 Unigramas:\n", top_unigrams)

# Bigramas
top_bigrams = top_ngrams(textos['text_content_anonymous'], ngram_range=(2,2))
print("\nTop 30 Bigramas:\n", top_bigrams)

# Trigramas
top_trigrams = top_ngrams(textos['text_content_anonymous'], ngram_range=(3,3))
print("\nTop 30 Trigramas:\n", top_trigrams)
```

###22. As 30 mensagens mais positivas (distintas);

```python
top_30_mensagens_positivas = executar_consulta("""
SELECT
    text_content_anonymous,
    score_sentiment
FROM fake_telegram
WHERE
    text_content_anonymous IS NOT NULL
    AND score_sentiment > 0
GROUP BY
    text_content_anonymous, score_sentiment
ORDER BY
    score_sentiment DESC
LIMIT 30;
""")

print(top_30_mensagens_positivas)
```

###23. As 30 mensagens mais negativas (distintas);

```python
# Consulta para as 30 mensagens mais negativas (distintas)
top_30_mensagens_negativas = executar_consulta("""
    SELECT
        text_content_anonymous,
        MIN(score_sentiment) AS score_sentiment
    FROM fake_telegram
    WHERE
        text_content_anonymous IS NOT NULL
        AND score_sentiment < 0
    GROUP BY
        text_content_anonymous, score_sentiment
    ORDER BY
        score_sentiment ASC
    LIMIT 30;
""")

top_30_mensagens_negativas
```

###29. As mensagens que possuem as palavras “FACÇÃO” e “CRIMINOSA”;

```python
# Consulta para encontrar mensagens com as palavras "FACÇÃO" e "CRIMINOSA", independentemente da capitalização
mensagens_faccao_criminosa = executar_consulta("""
    SELECT
        id_message,
        text_content_anonymous
    FROM fake_telegram
    WHERE
        LOWER(text_content_anonymous) LIKE '%facção%'
        AND LOWER(text_content_anonymous) LIKE '%criminosa%'
        AND text_content_anonymous IS NOT NULL;
""")

print(mensagens_faccao_criminosa)
```
